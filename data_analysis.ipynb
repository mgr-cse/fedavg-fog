```python
from dataSets import DataSet
import matplotlib.pyplot as plt

import numpy as np

dataset_iid = DataSet('mnist', 1)
dataset_niid = DataSet('mnist', 0)
```

```python
def plot_label_count(label: np.ndarray):
    counts = np.zeros(10, dtype=np.int32)
    for l in label:
        for i, e in enumerate(l):
            if e == np.float32(1.0): counts[i] += 1
    x = np.arange(len(counts))
    plt.figure(0)
    plt.bar(x, counts)
    plt.xticks(x)
    plt.xlabel('digits')
    plt.ylabel('frequency')
    plt.show()

# test dataset
plot_label_count(dataset_iid.train_label)

# train dataset
plot_label_count(dataset_iid.test_label)
```
